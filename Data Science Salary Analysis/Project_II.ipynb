```python
# Function to create a bar plot for average salary by a categorical variable
def plot_average_salary_by_category(category, data, rotation=0, figsize=(14, 7)):
    plt.figure(figsize=figsize)
    average_salary = data.groupby(category)['salary_in_usd'].mean().sort_values(ascending=False)
    sns.barplot(x=average_salary.index, y=average_salary.values, palette="coolwarm")
    plt.xticks(rotation=rotation)
    plt.title(f'Average Salary by {category.capitalize()} (2023)')
    plt.xlabel(category.capitalize())
    plt.ylabel('Average Salary in USD')
    plt.show()
```

```python
# the top 20 job titles by frequency
top_job_titles = data['job_title'].value_counts().head(20).index

# Now filter the data to include only these top job titles
top_jobs_data = data[data['job_title'].isin(top_job_titles)]

# Plotting average salary for these top job titles
plot_average_salary_by_category('job_title', top_jobs_data, rotation=90)
```
